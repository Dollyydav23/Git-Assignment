5. Tasks to Be Performed:
 1. Create a Git Flow workflow architecture on Git
 2. Create all the required branches
 3. starting from e feature branch, push the branch to the master, following the architecture
 4. Push an urgent.txt on master using hotfix

  - **Answer:** 1. Create a Git Flow workflow architecture on Git.
  
     Gitflow is a branching model for Git that organizes development workflows into structured branches, making it easier to manage feature development, releases, and hotfixes. It was introduced by Vincent Driessen and is particularly useful for projects with scheduled release cycles.

     It uses the following branches:

     Master/main:Production ready code.

     Develop: Latest development changes.

     feature/*: New features in progress.

     release/*: Code ready for final testing before production.

     hotfix/*: Urgent fixes on master/main.

     bugfix/*: Fixing bug that are not urgent.

    2. Creating all the required branches.

            --> git init
            --> echo "#Git flow project" > README.md
            --> git commit -m " Initialial commit on main"
            --> git branch main
           Now Creating develop branch for latest development changes.

            --> git checkout -b develop
            --> git push -u origin develop

          Creating feature branch for any new feature in progress.

            --> git checkout -b feature/login-page
            --> git push -u origin feature/login-page

          Creating release branch for final testing before production.

            --> git checkout develop
            --> git checkout -b release/v1.0
            --> git push -u origin release/v1.0

          Now finally creating a hotfix branch from main, if any urgent fixes needed.

            --> git checkout main
            --> git checkout -b hotfix/urgent-fix

    3. Starting from e feature branch, push the branch to the master, following the architecture

             --> git checkout feature/login-page
             --> echo " Login feature implemented." > login.txt
             Here we have add a new file named login.txt
             --> git add .
             --> git commit -m " Added login feature in feature branch."
             --> git push
             Now merging it in develop branch
             --> git checkout develop
             --> git merge feature/login-page -m "Merge login feature"
             --> git push

          Now we are moving from the develop branch to release branch which is used for final testing before production.

             --> git checkout release/v1.0
             --> git merge develop -m "Prepare release v1.0"
             Merge all the latest changes from the develop branch into release/v1.0
             --> git push

          Finalize the release by merging it into the main branch (production).Then, make sure any last-minute release changes (like bug fixes or version bumping) are also added back to develop.

            --> git checkout main
            --> git merge release/v1.0 -m "Release v1.0 merged to master"
            --> git push
            Also update develop with any final changes
            --> git checkout develop
            --> git merge release/v1.0 -m "Back-merge release v1.0 into develop"
            --> git push

    4. Push an urgent.txt on master using hotfix.

            Create hotfix branch from master/main.
            -->  git checkout main
            --> git checkout -b hotfix/fix-crash
            Now applying the fixes or any urgent fixes into the branches.
            --> echo "Urgent fix applied" > urgent.txt
            --> git add urgent.txt
            --> git commit -m "Apply urgent fix for production issue"
            --> git push -u origin hotfix/fix-crash

        now  merging the hotfix to main/master.
          
             --> git checkout main
             --> git merge hotfix/fix-crash -m "Merge urgent hotfix"
             --> git push

        Finally, Back-merge hotfix to develop

             --> git checkout develop
             --> git merge hotfix/fix-crash -m "Back-merge hotfix to develop"
             --> git push
